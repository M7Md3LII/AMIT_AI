## Try to make a full EDA with visualizations to investigate the data.

```python
#the Age in 20,0.030 is Height
sns.distplot(df['Age']);
```

```python
#the Height is on Survived (20-50),dead(20-50)
sns.jointplot(x='Age',y='Survived',data=df,kind='scatter');
```

```python
#specifically the dead betwen ( 20 - 30 ) , the Survived betwen ( 20 - 40 )
sns.jointplot(x='Age',y='Survived',data=df,kind='hex')
```

```python
#the number of ages increases , the survival rate decreases
sns.jointplot(x='Age',y='Survived',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
plt.subplots(figsize=(17,7))
plt.xticks(rotation=90)
sns.barplot(x='Age',y='Survived',data=df);
```

```python
plt.subplots(figsize=(17,7))
plt.xticks(rotation=90)
sns.boxplot(x="Age", y="Survived", data=df,palette='rainbow');
```

```python
plt.subplots(figsize=(10,7)) #to zoom or big size the outpot
sns.boxplot(data=df,palette='rainbow',orient='h');
#Fare have too many out lire reverse the rest
```

```python
sns.pairplot(df,hue='Fare',palette='coolwarm');
```

```python
plt.subplots(figsize=(11,10)) #to zoom or big size the outpot
sns.heatmap(df.corr()); # dark degree direct relationship 
                        # light degree is an inverse relationship
                        # (pclass - Fare) is an a dark degree direct relationship
                        
```

```python
plt.subplots(figsize=(11,10)) #to zoom or big size the outpot
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)# blue degree direct relationship 
                                                 # light degree is an inverse relationship
                                                 # (pclass - Fare) is an a blue degree direct relationship by (-0.55)
```
